```python
import pandas as pd
import joblib
import category_encoders as ce
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# 1. Load Dataset
data = pd.read_csv("House_Rent_Dataset.csv")

# 2. Separate Features and Target
X = data.drop(columns=["Rent", "Posted On", "Floor", "Area Locality"])
y = data["Rent"]

# 3. Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# 4. Build the Pipeline (This glues everything together!)
pipeline = Pipeline([
    ("encoder", ce.LeaveOneOutEncoder()),
    ("scaler", MinMaxScaler()),
    ("model", LinearRegression())
])

# 5. Train the entire Pipeline at once
pipeline.fit(X_train, y_train)

# 6. Save the entire Pipeline into a single .pkl file
joblib.dump(pipeline, "house_rent_pipeline.pkl")
print("Pipeline saved successfully as house_rent_pipeline.pkl!")
```

```python
import pandas as pd
data = pd.read_csv("wine.csv")
y = data['quality']
x = data.drop('quality', axis=1)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state=42)
```

```python
from sklearn.neural_network import MLPClassifier
model = MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=500)
model.fit(x_train,y_train)

print(model.score(x_test,y_test))
```

```python
from sklearn.metrics import confusion_matrix,classification_report
y_pred=model.predict(x_test)
confusion_matrix=confusion_matrix(y_test,y_pred)
print(confusion_matrix)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model=Sequential()
model.add(Dense(64,input_dim=x_train.shape[1],activation='relu'))
model.add(Dense(6,activation='softmax'))
model.summary()
```
